```python
import qiskit
from qiskit import*
from qiskit.tools.visualization import*
```

```python
circuit=QuantumCircuit(3,3)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
# creating entanglement between q1 and q2

circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

```python
# teleporting 

circuit.cx(0,1)
circuit.h(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.measure([0,1],[0,1])
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

state 1 is teleported from q0 to q2

#### verifying by measuring q2


```python
circuit.measure(2,2)
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,backend=simulator,shots=1024).result()
counts=result.get_counts()
plot_histogram(counts)
```

 look ^ q2 is always 1
